```python
from pprint import pprint
n = 5
board = [['.']*n for i in range(n)]
pprint(board)
```

```python
board[0][0] = 'S'
board[0][3] = 'X'
board[1][1] = 'X'
board[1][3] = 'X'
board[2][1] = 'X'
board[3][2] = 'X'
board[3][3] = 'X'
board[4][4] = 'G'
pprint(board)
```

```python
def is_safe(board, row, col):
    
    if 0 <= row < len(board) and 0 <= col < len(board[0]):
        if board[row][col] == 'X':
            return False
        
        return True
    
    return False
```

```python
def moves(board, row, col):
    
    possible_moves = []
    
    # Right move
    if is_safe(board, row, col+1):
        possible_moves.append((row, col+1))
        
    # Left move
    if is_safe(board, row, col-1):
        possible_moves.append((row, col-1))
        
    # Up move
    if is_safe(board, row-1, col):
        possible_moves.append((row-1, col))
        
    # Down move
    if is_safe(board, row+1, col):
        possible_moves.append((row+1, col))
    
    return possible_moves
```

```python
from collections import deque

def bfsSearch(board):
    initial_state = (0,0)
    
    queue = deque()
    queue.append((initial_state,[]))
    
    visited = set()
    visited.add(initial_state)
    
    while queue:
        current, path = queue.popleft()
        r, c = current
        
        if board[r][c] == 'G':
            return path
    
        for move in moves(board, r, c):
            
            if move not in visited:
                visited.add(move)
                
                queue.append(
                    (move, path + [move])
            )
```

```python
solution = bfsSearch(board)
for move in solution:
    board[move[0]][move[1]] = '#'

pprint(board)
```
