```python
from collections import deque

def is_valid(state):
    A, B = state
    
    if 0 <= A <= 4 and 0 <= B <= 3:
        return True
    
    return False
```

```python
def actions(state):
    A, B = state
    moves = []
    
    # Fill A
    new_state = (4, B)
    moves.append(("Fill jug A", new_state))
    
    # Fill B
    new_state = (A, 3)
    moves.append(("Fill jug B", new_state))
    
    # Empty jug A
    new_state = (0, B)
    moves.append(("Empty jug A", new_state))
    
    # Empty jug B
    new_state = (A, 0)
    moves.append(("Empty jug B", new_state))
    
    # Pour From jug A to B
    if A > 0 and B < 3:
        transfer = min(A, 3-B)
        new_state = (A - transfer, B + transfer)
        moves.append(("Pour from jug A to B", new_state))
        
    # Pour From Jug B to A
    if B > 0 and A < 4:
        transfer = min(B, 4-A)
        new_state = (A + transfer, B - transfer)
        moves.append(("Pour from jug B to A", new_state))
        
    return moves
```

```python
def bfs():
    initial_state = (0, 0)
    
    queue = deque()
    queue.append((initial_state,[]))
    
    visited = set()
    visited.add(initial_state)
    
    while queue:
        current, path = queue.popleft()
        A, B = current
        if A == 2:
            return path
        
        for action, new_state in actions(current):
            if is_valid(new_state) and new_state not in visited:
                visited.add(new_state)
                
                queue.append(
                    (new_state, path + [action] )
                )
                
    return None
```

```python
solution = bfs()
for step in solution:
    print(step)
```
