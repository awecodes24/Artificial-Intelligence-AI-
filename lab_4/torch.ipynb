```python
import torch

# 1. Check for GPU (Your system prints: "Using device= cuda")
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device= {device}")

# 2. Create a random tensor (defaults to CPU memory)
cpu_tensor = torch.rand(3, 3)
print(f"Original location: {cpu_tensor.device}")

# 3. Explicitly move the tensor to your RTX 4050 GPU
gpu_tensor = cpu_tensor.to(device)
print(f"New location: {gpu_tensor.device}")

# 4. Create a neural network model and move it to the GPU
model = torch.nn.Linear(3, 1).to(device)
print(f"Model parameters are on: {next(model.parameters()).device}")
```
